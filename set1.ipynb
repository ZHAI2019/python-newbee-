```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

df = pd.read_excel(r"D:\jupy nt\2019_6_datahomework\set3\result\table.xlsx",sheetname = "LiYao_Datare")
df = pd.DataFrame(df)
df.head()


data = df[['s1-s2','Age']]

data1 = data["s1-s2"]
data2 = data["Age"]
a = np.array(data1)
b = np.array(data2)
list1 = a.tolist()
list2 = b.tolist()


r = np.corrcoef(list1,list2)[0, 1]
print(r)
a = list(stats.pearsonr(data['s1-s2'],data["Age"]))
print(a)
r = a[0]
p = a[1]
plt.title(" $set1(s1-s2)× age$")
plt.xlabel(" Age      " + " r = " + str('%6f'%r) + "  $p$ = " + str('%6f'%p))
plt.ylabel("s1-s2")
plt.scatter(list2,list1)

plt.savefig(r"D:\jupy nt\2019_6_datahomework\set1\result\ $set1(s1-s2)× age$.png")
plt.show()
```
